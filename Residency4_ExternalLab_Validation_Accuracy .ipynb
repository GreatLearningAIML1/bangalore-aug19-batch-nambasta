### Q2.Prepare the model data by converting non-numeric to dummy ( 1 Marks)
##### Hint: Use get_dummies

```python
# Print Shape of model data
gcredit_df.shape
```

```python
for feature in gcredit_df.columns: # Loop through all columns in the dataframe
    if gcredit_df[feature].dtypes == 'object': # Only apply for columns with categorical strings
        temp = pd.get_dummies(gcredit_df[feature],prefix=feature)
        gcredit_df = gcredit_df.join(temp)

# gcredit_new_df = pd.get_dummies(gcredit_df,prefix='dummy',drop_first=True)
```

### Drop the original variables which are converted to dummy

```python
for feature in gcredit_df.columns: # Loop through all columns in the dataframe
    if gcredit_df[feature].dtypes == 'object': # Only apply for columns with categorical strings
        gcredit_df.drop(gcredit_df[[feature]],axis=1,inplace = True)
```

### Q3 Split Train/Test data 70:30 ratio( 1 Marks)
##### Hint:from sklearn.model_selection import train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
test_size = 0.30
X = gcredit_df.iloc[:,1:63]
y = gcredit_df.iloc[:,0]
```

```python
X.info()
```

```python
y.head()
```

```python
seed =1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

### Q4 Build Random Forest Model( 1 Marks)
#### Hint:from sklearn.ensemble import RandomForestClassifier using n_jobs=2,n_estimators=500,criterion="entropy",random_state=9999

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=500,n_jobs=2,criterion='entropy',random_state=9999)
```

```python
from sklearn import preprocessing

# scale all the columns of the mpg_df. This will produce a numpy array
#pima_df_scaled = preprocessing.scale(pima_df[0:7])
X_train_scaled = preprocessing.scale(X_train)
X_test_scaled = preprocessing.scale(X_test)
```

```python
model.fit(X_train_scaled, y_train)
```

```python
y_predict = model.predict(X_test_scaled)
```

### Q5 Calculate Confusion Matrix and Accuracy score (1 Marks)
##### Hint: Use confusion_matrix and accuracy_score

```python
model.score(X_test_scaled , y_test) 
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test,y_predict)

conf_matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```

### Q7 K-fold cross-validation( 2 Marks)
##### k-fold cross validation( without stratification)
##### Usually k is set as 10-20 in practical settings, depends on data set size

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
# Use below values
num_folds = 10
seed = 77
kfold = KFold(n_splits=num_folds, random_state=seed)
```

```python
#Validate the Random Forest model build above using k fold
```

```python
results = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=kfold)
print("Mean Square Error: %.3f (%.3f)" % (-results.mean(), results.std()))
```

```python
#Calculate Mean score
```

```python
results = cross_val_score(model, X, y, scoring='r2', cv=kfold)
print("Mean Square Error: %.3f (%.3f)" % (-results.mean(), results.std()))
```
