```python
#1.Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.
import numpy as np
import pandas as pd
#Reading data set
laliga_orig = pd.read_csv('Laliga.csv',skiprows=1)
#Replacing dashes with 0.
laliga = laliga_orig.replace('-',0)
#Output the result
print(laliga)
```

```python
#2.Print all the teams which have started playing between 1930-1980.
print(laliga[(laliga.Debut > '1930') & (laliga.Debut <'1980')][['Team','Debut']])
```

```python
#3.Print the list of teams which came Top 5 in terms of points

#Create a deep copy so that original data set remains unaffected.
df = laliga.copy()

#Convert Points (dtype=Object) to numeric.
df['Points'] = pd.to_numeric(df['Points'])

#Sortin descending order based on 'Points' and print the top 5 teams.
df = df.sort_values('Points',ascending=False)
print(df[['Team','Points']].head(5))
```

```python
#4.Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
   #Using the same function, find the team which has maximum and minimum goal difference.
    
#Function that returns team wise goal differences  
def Goal_diff_count(laliga):
    laliga['GoalsFor'] = pd.to_numeric(laliga['GoalsFor'])
    laliga['GoalsAgainst'] = pd.to_numeric(laliga['GoalsAgainst'])
    laliga['Goal_diff'] = laliga['GoalsFor'] - laliga['GoalsAgainst']
    temp = laliga[['Team','GoalsFor','GoalsAgainst','Goal_diff']]
    return temp

#Outputs using 'Goal_diff_count' function
print("Dataframe with Goal Differences::::\n\n",Goal_diff_count(laliga),"\n\n")

print("Team with maximum goal difference::::\n\n",Goal_diff_count(laliga).loc[Goal_diff_count(laliga)['Goal_diff'].idxmax()],"\n\n")

print("Team with minimum goal difference::::\n\n",Goal_diff_count(laliga).loc[Goal_diff_count(laliga)['Goal_diff'].idxmin()])
        
```

```python
#5.Create a new column with name “Winning Percent” and append it to the data set.
   #If there are any numerical error, replace it with 0%
   #Print the top 5 teams which has the highest Winning percentage.

#Function that returns Winning percentage of the teams.
def Get_Data_With_WinningPercentage(laliga_data):
    df = laliga_data.copy()
    df['GamesWon'] = pd.to_numeric(df['GamesWon'])
    df['GamesPlayed'] = pd.to_numeric(df['GamesPlayed'])
    
    #Calculate Winning Percentage
    df['Winning Percent'] = (df['GamesWon'] / df['GamesPlayed']) * 100
    
    #Fill in missing data.
    df_clean = df[['Team','GamesPlayed','GamesWon','Winning Percent']].fillna(0)
    
    print("Dataframe with 0% \n\n",df_clean,"\n\n")
    
    return df_clean

#Printing top 5 teams w.r.t their winning percentages
print("Top 5 teams with highest Winning percentage::\n\n",Get_Data_With_WinningPercentage(laliga).sort_values('Winning Percent',ascending=False).head(5))
    
```

```python
#6.Group teams based on their “Best position” and print the sum of their points for all positions

laliga['Points'] = pd.to_numeric(laliga['Points'])
df = laliga.groupby('BestPosition')
print(df[['Points']].sum())
```
