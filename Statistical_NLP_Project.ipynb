```python
#Reducing the sample size by choosing few topics.

blogs_student = blogs[ (blogs.topic == 'Student') ] 
blogs_tech = blogs[ (blogs.topic == 'Technology') ]
blogs_arts = blogs[(blogs.topic == 'Arts')]       
blogs_non_profit = blogs[(blogs.topic == 'Non-Profit')]
blogs_law = blogs[blogs.topic == 'Law']
blogs_gov = blogs[(blogs.topic == 'Government')]
blogs_rel = blogs[(blogs.topic == 'Religion')]
blogs_fashion = blogs[(blogs.topic == 'Fashion')]
```

```python
#Downsampling- To remove data imbalances
from sklearn.utils import resample

blogs_student = resample(blogs_student, 
                      replace=True,    # sample with replacement
                      n_samples=2000,     # to match number of values in each class
                      random_state=123) # reproducible results

blogs_tech = resample(blogs_tech, 
                      replace=True,    # sample with replacement
                      n_samples=2000,     # to match number of values in each class
                      random_state=123) # reproducible results

blogs_arts = resample(blogs_arts, 
                      replace=True,    # sample with replacement
                      n_samples=2000,     # to match number of values in each class
                      random_state=123) # reproducible results

blogs_non_profit = resample(blogs_non_profit, 
                      replace=True,    # sample with replacement
                      n_samples=2000,     # to match number of values in each class
                      random_state=123) # reproducible results

blogs_law = resample(blogs_law, 
                      replace=True,    # sample with replacement
                      n_samples=2000,     # to match number of values in each class
                      random_state=123) # reproducible results

blogs_gov = resample(blogs_gov, 
                      replace=True,    # sample with replacement
                      n_samples=2000,     # to match number of values in each class
                      random_state=123) # reproducible results

blogs_rel = resample(blogs_rel, 
                      replace=True,    # sample with replacement
                      n_samples=2000,     # to match number of values in each class
                      random_state=123) # reproducible results

blogs_fashion = resample(blogs_fashion, 
                      replace=True,    # sample with replacement
                      n_samples=2000,     # to match number of values in each class
                      random_state=123) # reproducible results
```

```python
blogs = pd.concat([blogs_student, blogs_tech, blogs_arts, blogs_non_profit, blogs_law, blogs_gov, blogs_rel, blogs_fashion ])
```

```python
 # DATA PREPROCESSING STARTS
```

```python
#Changing to lowercase
blogs['text'] = blogs['text'].apply( lambda t : ' '.join( word.lower() for word in t.split()  ) )
```

```python
blogs['text'].head(5)
```

```python
#Remove unwanted characters
blogs['text'] = blogs['text'].str.replace( '[^\w\s]' , '' )
blogs['text'].head(10)
```

```python
#Remove stop words
import nltk
nltk.download('stopwords')
```

```python
stop = stopwords.words('english')
blogs['text'] = blogs['text'].apply( lambda t : " ".join( word for word in t.split() 
                                                      if word not in stop ) )
blogs['text'].head(10)
```

```python
#Remove unwanted spaces
blogs['text'] = blogs['text'].str.strip()
blogs['text'].head(10)
```

```python
blogs.age = blogs.age.astype(str)
```

```python
##Merge all the label columns together
blogs['labels'] = blogs[['gender', 'age', 'topic', 'sign']].values.tolist()
blogs.head()
```

```python
blogs.dtypes
```

```python
blogs_final = blogs[['text','labels']]
blogs_final.head()
```

```python
blogs_final.shape
```

```python
blogs_final.columns
```

```python
#Separate features and labels
x = blogs_final['text']
y = blogs_final['labels']
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
blogs_final.shape
```

```python
# Vectorize the features -------------------
```

```python
#Create Bag of Words using count vectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(1,2))
```

```python
#Vectorize training and testing features 
x_train_dtm = cv.fit_transform(x_train)
x_test_dtm = cv.transform(x_test)
```

```python
print('Features: ', x_train_dtm)
```

```python
#Create a dictionary to get the count of every label
blogs['labels_string'] = blogs[['gender', 'age', 'topic', 'sign']].apply(lambda x: ','.join(x), axis = 1)
cv_train_label = CountVectorizer(analyzer = 'word')
cv_train_label.fit_transform(blogs['labels_string'])
cv_train_label.vocabulary_
```

```python
##Transform labels using MultiLabelBinarizer ----
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
```

```python
y_train_transformed = mlb.fit_transform(y_train)
```

```python
y_test_transformed = mlb.fit_transform(y_test)
```

```python
mlb.classes_
```

```python
y_train_transformed
```

```python
y_test_transformed
```

```python
#Building model
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(solver='lbfgs',max_iter = 1500)
clf = OneVsRestClassifier(clf)
```

```python
clf.fit(x_train_dtm, y_train_transformed)
```

```python
y_pred = clf.predict(x_test_dtm)
```

```python
#Print scores ::::::::::

from sklearn import metrics
print(metrics.accuracy_score(y_test_transformed, clf.predict(x_test_dtm)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test_transformed, clf.predict(x_test_dtm)))
```
