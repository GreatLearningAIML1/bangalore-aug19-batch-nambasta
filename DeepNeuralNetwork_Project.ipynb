```
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
#cat_cols = ['Geography','Gender']
data['Geography'] = labelencoder.fit_transform(data['Geography'])
data['Gender'] = labelencoder.fit_transform(data['Gender'])
```

```
data.info()
```

```
from sklearn.model_selection import train_test_split

#Distinguish the feature and target set
X =  data.drop("Exited", axis=1)
y = data.pop("Exited")

# splitting data into training and test set for independent attributes
train_set, test_set, train_labels, test_labels = train_test_split(X, y, test_size=0.30, random_state=1)
```

```
test_set.info()
```

```
# Import `StandardScaler` from `sklearn.preprocessing`
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#Normalize all numeric columns.
num_cols = train_set.columns[train_set.dtypes.apply(lambda c: np.issubdtype(c, np.number))]
train_set[num_cols] = scaler.fit_transform(train_set[num_cols])

# Scale the test set
test_set[num_cols] = scaler.fit_transform(test_set[num_cols])
```

```
# Import `Sequential` from `keras.models`
from keras.models import Sequential

# Import `Dense` from `keras.layers`
from keras.layers import Dense

# Initialize the constructor
model = Sequential()

# Add an input layer 
model.add(Dense(10, activation='sigmoid', input_shape=(10,)))

# Add one hidden layer 
model.add(Dense(8, activation='sigmoid'))

# Add an output layer 
model.add(Dense(1, activation='sigmoid'))
```

```
#model.compile(loss='squared_hinge',
#model.compile(loss='hinge',
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
                   
#history = model.fit(train_set, test_set,epochs=20, batch_size=1, verbose=1)
```

```
model.summary()
```

```
model.fit(train_set,train_labels,validation_data=(test_set,test_labels),epochs=20)
```

```
#Predict by default uses 0.5 threshold
y_pred = np.round(model.predict(test_set))
y_pred
```

```
#Score
score = model.evaluate(test_set, test_labels,verbose=1)
print(score)
```

```
from sklearn import metrics
metrics.confusion_matrix(test_labels,y_pred)
```
